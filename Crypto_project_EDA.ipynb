# Importing The libraries used in the EDA process And uploading the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path = 'D:\לימודים,\crypto-project-summer-2024/Bitcoin-Historical-Data.xlsx'
df_bitcoin = pd.read_excel(file_path, sheet_name='Bitcoin Historical Data (1)')

# Display the first few rows of the data
df_bitcoin.head()
```

# Desplaying the type of data in each column and amount of entries
## And desplaying  if any column has missing values

```python
df_bitcoin.info()

# Checking for missing values
df_bitcoin.isnull().sum()
```

# Handling missing data with Backward fill 

```python
# Apply backward fill to fill missing values
df_bitcoin.fillna(method='bfill', inplace=True)

# Display the first few rows of the dataset after handling missing data
df_bitcoin.head()
```

# Summary of Statistical Data

```python
df_bitcoin.describe()
```

```python
 #Convert 'Vol.' and 'ETH Vol.' to numeric after removing 'K' and 'M'
df_bitcoin['Vol.'] = pd.to_numeric(df_bitcoin['Vol.'].str.replace('K', 'e3').str.replace('M', 'e6'), errors='coerce')
df_bitcoin['ETH Vol.'] = pd.to_numeric(df_bitcoin['ETH Vol.'].str.replace('K', 'e3').str.replace('M', 'e6'), errors='coerce')
```

# Visualization of Bitcoin based on the price over time

```python
# Plotting Bitcoin Price over time
plt.figure(figsize=(10,6))
plt.plot(df_bitcoin['DATE'], df_bitcoin['Price'], label='Bitcoin Price', color='blue')
plt.title('Bitcoin Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

# Visualization of Distribution of Percentile change in Bitcoin price

```python
# Plotting the distribution of Change %
plt.figure(figsize=(10,6))
plt.hist(df_bitcoin['Change %'], bins=50, color='purple')
plt.title('Distribution of Percentage Change in Bitcoin Price')
plt.xlabel('Percentage Change')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
corr_matrix = df_bitcoin.corr()
```

```python
# Set the figure size for better readability
plt.figure(figsize=(12, 8))

# Plotting a heatmap of the correlations
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Add a title to the heatmap
plt.title('Correlation Heatmap of Bitcoin Data')

# Show the plot
plt.show()
```

# Visualization of the outliers in the price of Bitcoin

```python
# Plotting a boxplot for a single variable (e.g., 'Price')
plt.figure(figsize=(10, 6))
plt.boxplot(df_bitcoin['Price'])
plt.title('Boxplot of Bitcoin Price')
plt.show()
```

```python
# Calculate IQR for a column (e.g., 'Price')
Q1 = df_bitcoin['Price'].quantile(0.25)
Q3 = df_bitcoin['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_without_outliers = df_bitcoin[(df_bitcoin['Price'] >= lower_bound) & (df_bitcoin['Price'] <= upper_bound)]

# Show the cleaned data
df_without_outliers.head()
```

```python
# Capping the outliers at lower and upper bounds
df_bitcoin['Price'] = df_bitcoin['Price'].clip(lower=lower_bound, upper=upper_bound)

# Show the modified data
df_bitcoin.head()
```

```python
df_bitcoin['Price_log'] = np.log(df_bitcoin['Price'])
```

```python
# Step 1: Calculate IQR (Interquartile Range) for the 'Price' column
Q1 = df_bitcoin['Price'].quantile(0.25)  # First quartile (25th percentile)
Q3 = df_bitcoin['Price'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile Range

# Step 2: Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR  # Lower boundary for outliers
upper_bound = Q3 + 1.5 * IQR  # Upper boundary for outliers

# Step 3: Filter out the rows where 'Price' values are outside the IQR boundaries
df_without_outliers = df_bitcoin[(df_bitcoin['Price'] >= lower_bound) & (df_bitcoin['Price'] <= upper_bound)]

# Step 4: Display the cleaned data
df_without_outliers.head()  # Show the first few rows of the data without outliers
```

```python
# Plot the trend of Bitcoin price over time after removing outliers
plt.figure(figsize=(12, 6))

# Plot Bitcoin Price over time
plt.plot(df_without_outliers['DATE'], df_without_outliers['Price'], label='Bitcoin Price', color='blue')

# Add labels and title
plt.title('Bitcoin Price Trend Over Time (After Removing Outliers)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
